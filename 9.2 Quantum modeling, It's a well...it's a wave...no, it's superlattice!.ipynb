## 1. Full Code

```python
__author__ = 'J11'


import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eig

h = 1
m = 1
length = 1
potentials = [10, 0]
nodes = 100

def create_potential_grid(potentials, nodes):
    n = int(nodes/len(potentials))
    v = []
    for pot in potentials:
        for i in range(n):
            v.append(pot)
    return np.array(v)

def create_hamiltonian_matrix(potential, wavevector, dx):
    v = potential
    N = len(v)
    k = wavevector
    H = np.zeros((N, N), dtype=complex)
    a = -h**2/2/m
    H1 = np.array([1/dx**2-1.0j*k/dx, -2/dx**2-k**2, 1/dx**2+1.0j*k/dx])*a
    for n in range(N)[1:-1]:
        H[n, n-1:n+2] = H1
    #first row
    H[0, 0:2] = np.array([-2/dx**2-k**2, 1/dx**2+1.0j*k/dx])*a
    H[0, -1] = (1/dx**2-1.0j*k/dx)*a
    #last row
    H[-1, -2:] = np.array([1/dx**2-1.0j*k/dx, -2/dx**2-k**2])*a
    H[-1, 0] = (1/dx**2+1.0j*k/dx)*a
    H += np.diag(v)
    return H

v = create_potential_grid(potentials, nodes)
k_list = np.linspace(-1, 1.2*np.pi/length, 101)

nbands = 5
eng = np.zeros((nbands, len(k_list)))

for i,k in enumerate(k_list):
    H = create_hamiltonian_matrix(v, k, dx=length/nodes)
    eigvals, eigvecs = eig(H)
    ind_ord = np.argsort(eigvals)
    eigvals = eigvals[ind_ord]
    eigvecs = eigvecs[ind_ord]
    print(k)
    for j in range(nbands):
        eng[j, i] = eigvals[j]

fig = plt.figure(facecolor='w')
for j in range(nbands):
    a, = plt.plot(k_list, eng[j, :], linewidth = 2, label='Band ' + str(j+1))
    plt.legend(loc='best')
    plt.axvline(x=np.pi/length, ymin=0.0, ymax = np.max(eng), linewidth=1, color='k')
    plt.axvline(x=0, ymin=0.0, ymax = np.max(eng), linewidth=1, color='k')
    plt.yticks([])
    plt.xlabel('Wavevector')
    plt.ylabel('Energy a.u')
    plt.grid()
plt.show()
```
